## Question 8 (2 points)

Estimate AR(p) and MA(q) models for p = 1,...,24 and q = 1,...,24 

using the calculated monthly growth rates of CPI. 

Include a constant in the models. 

Select which model you would chose based on the values of the Akaike Information Criterion (AIC).

```python
# Add a constant to the DataFrame df_growth
df_growth['const'] = 1
df_growth.head()
```

```python
# create a list of p and q values
values = range(1, 25)

# create an empty arrays to store the results
results_ar_aic = []
results_ma_aic = []

# loop over all p and q values and fit the ARMA model
for val in values:
    model = sm.tsa.ARIMA(df_growth['CPI_growth'].dropna(), order=(val, 0, 0), trend='c')
    res = model.fit()
    results_ar_aic.append(res.aic)

    model = sm.tsa.ARIMA(df_growth['CPI_growth'].dropna(), order=(0, 0, val), trend='c')
    res = model.fit()
    results_ma_aic.append(res.aic)
    
```

```python
# Store the results in a DataFrame
df_results_ar_aic = pd.DataFrame({'p': values, 'AIC': results_ar_aic})
df_results_ma_aic = pd.DataFrame({'q': values, 'AIC': results_ma_aic})

# Merge the two DataFrames
df_results_aic = pd.merge(df_results_ar_aic, df_results_ma_aic, left_on='p', right_on='q')

# Rename columns as p, AIC_p, q, AIC_q
df_results_aic.columns = ['p', 'AIC_p', 'q', 'AIC_q']
```

```python
# Find the lowest value of AIC_q and the corresponding q
q_star = df_results_aic['q'][df_results_aic['AIC_q'] == df_results_aic['AIC_q'].min()].values[0]

# Find the lowest value of AIC_p and the corresponding p
p_star = df_results_aic['p'][df_results_aic['AIC_p'] == df_results_aic['AIC_p'].min()].values[0]

# Print the estimated values of p_star and q_star and their corresponding AIC values
print('best p:', p_star, ' AIC_p:', df_results_aic['AIC_p'].min())
print('best q:', q_star, ' AIC_q:', df_results_aic['AIC_q'].min())

if p_star < q_star:
    print('The best model is an AR(%d) model with an AIC of %0.2f.' % (p_star, df_results_aic['AIC_p'].min()))
else:
    print('The best model is an MA(%d) model with an AIC of %0.2f.' % (q_star, df_results_aic['AIC_q'].min()))
```

In general, a lower value of AIC (Akaike Information Criterion) indicates a better model fit. AIC is a criterion that is used to compare different statistical models and select the one that best explains the data. It is defined as:

AIC = 2k - 2ln(L)

where k is the number of parameters in the model, and L is the likelihood function of the data given the model.

The AIC penalizes models that have more parameters, so a lower value indicates a model that fits the data well with fewer parameters. Therefore, when comparing different models, the one with the lowest AIC is usually preferred.

Inspect the properties of the time series residuals of the AR(9) model. 

Are the residuals following a White Noise process? Are they autocorrelated? 

Are they homoskedastic? 

You can address these questions by calculating 

mean 

and standard deviation, 

calculating the autocorrelation function based on the residuals, 

and plotting the residuals. 

What conclusions do you draw with respect to the adequacy of the estimated AR(9) model?

```python
# obtain residuals
residuals = results_ar9.resid

# calculate mean and standard deviation of residuals
residuals_mean = np.mean(residuals)
residuals_std = np.std(residuals)

print('Mean of residuals:', residuals_mean)
print('Standard deviation of residuals:', residuals_std)

# plot histogram of residuals
plt.hist(residuals, bins=20, density=True, alpha=0.6)
x = np.linspace(residuals_mean - 3*residuals_std, residuals_mean + 3*residuals_std, 100)
plt.plot(x, norm.pdf(x, residuals_mean, residuals_std), 'r-', lw=2)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Density')
plt.show()

# plot autocorrelation function of residuals
sm.graphics.tsa.plot_acf(residuals, lags=48)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function of Residuals')
plt.show()

# plot residuals over time
plt.plot(residuals)
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.title('Residuals Over Time')
plt.show()
```

Residuals are said to follow a white noise process if they

-are independent, identically distributed

-have constant mean and variance over time. 

In other words, there should be no discernible pattern in the residuals, 

and they should not be related to each other in any way. 

If the residuals follow a white noise process, 

it suggests that the model has captured all the information in the data 

and there is no additional structure left to explain.

### AR(1)

```python
# Create df_ar1 by copying CPI and const of df, then add a column of lag 1
df_ar1 = df_growth[['CPI_growth', 'const']].copy()

# Add a column of lag 1
df_ar1['CPI_lag1'] = df_growth['CPI_growth'].shift(1)

# Drop missing values
df_ar1.dropna(inplace=True)

df_ar1
```

```python
# Get argument index for the start of the forecasting period
start_index = df_ar1.index.get_loc('2015-01-01')

print(start_index)

print(len(df_ar1))

df_ar1[:start_index]
```

```python
def forecasting_ar1(n_step, df):
    forecast_ar1_step= []

    for i in range(start_index, len(df)):
        estimation_window = df[i-start_index:i+1-n_step]
        # # Fit OLS model
        model_ar1 = sm.OLS(estimation_window['CPI_growth'], estimation_window[['const', 'CPI_lag1']])
        results_ar1 = model_ar1.fit()
        # # Get the coefficient for the lagged variable
        phi_ar1 = results_ar1.params['CPI_lag1']
        const_ar1 = results_ar1.params['const']
        # Make n-step ahead forecast
        prediction = prediction_next(const_ar1, phi_ar1,estimation_window['CPI_growth'][-1])
        if (n_step > 1):
            for i in range(n_step-1):
                prediction = prediction_next(const_ar1, phi_ar1, prediction)
        forecast_ar1_step.append(prediction)

    return forecast_ar1_step

def prediction_next(const, beta, y_lag):
    return const + beta * y_lag

def df_forecast_ar1():
    forecast_ar1_step1 = forecasting_ar1(1, df_ar1)
    forecast_ar1_step2 = forecasting_ar1(2, df_ar1)
    forecast_ar1_step3 = forecasting_ar1(3, df_ar1)
    forecast_ar1_step4 = forecasting_ar1(4, df_ar1)

    # Make a dataframe of the forecast of AR 1, name coulms AR(1) step: 1, AR(1) step: 2, AR(1) step: 3, AR(1) step: 4
    df_ar1_results = pd.DataFrame({'AR(1) step: 1': forecast_ar1_step1,
    'AR(1) step: 2': forecast_ar1_step2,
    'AR(1) step: 3': forecast_ar1_step3,
    'AR(1) step: 4': forecast_ar1_step4}, index=df_ar1[start_index:].index)

    # Add real values
    df_ar1_results['True data'] = df_ar1.CPI_growth[start_index:]

    return df_ar1_results

def computeMSE(df):
    mse_ar1 = []
    for col in df.columns[:-1]:
        mse = mean_squared_error(df[col], df['True data'])
        mse_ar1.append(mse)

    # Print MSE for each column
    for i, col in enumerate(df.columns[:-1]):
        print(f"MSE for {col}: {mse_ar1[i]}")

    return mse_ar1
```

```python
def forecasting_ar9(n_step, df):
    forecast_ar9_step= []

    for i in range(start_index, len(df)):
        estimation_window = df[i-start_index:i+1-n_step]
        # # Fit OLS model
        model_ar9 = sm.OLS(estimation_window['CPI_growth'], estimation_window[['const', 'CPI_lag1', 'CPI_lag2', 'CPI_lag3', 'CPI_lag4', 'CPI_lag5', 'CPI_lag6', 'CPI_lag7', 'CPI_lag8', 'CPI_lag9']])
        results_ar9 = model_ar9.fit()
        # # Get the coefficient for the lagged variable
        const_ar = results_ar9.params['const']
        phi1 = results_ar9.params['CPI_lag1']
        phi2 = results_ar9.params['CPI_lag2']
        phi3 = results_ar9.params['CPI_lag3']
        phi4 = results_ar9.params['CPI_lag4']
        phi5 = results_ar9.params['CPI_lag5']
        phi6 = results_ar9.params['CPI_lag6']
        phi7 = results_ar9.params['CPI_lag7']
        phi8 = results_ar9.params['CPI_lag8']
        phi9 = results_ar9.params['CPI_lag9']
        # Make n-step ahead forecast
        prediction = prediction_next(const_ar, phi9, phi8, phi7, phi6, phi5, phi4, phi3, phi2, phi1, estimation_window['CPI_lag1'][-1], estimation_window['CPI_lag2'][-1], estimation_window['CPI_lag3'][-1], estimation_window['CPI_lag4'][-1], estimation_window['CPI_lag5'][-1], estimation_window['CPI_lag6'][-1], estimation_window['CPI_lag7'][-1], estimation_window['CPI_lag8'][-1], estimation_window['CPI_lag9'][-1])
        if (n_step > 1):
            for i in range(n_step-1):
                prediction = prediction_next(const_ar, phi9, phi8, phi7, phi6, phi5, phi4, phi3, phi2, phi1, prediction, estimation_window['CPI_lag1'][-1], estimation_window['CPI_lag2'][-1], estimation_window['CPI_lag3'][-1], estimation_window['CPI_lag4'][-1], estimation_window['CPI_lag5'][-1], estimation_window['CPI_lag6'][-1], estimation_window['CPI_lag7'][-1], estimation_window['CPI_lag8'][-1])
        forecast_ar9_step.append(prediction)

    return forecast_ar9_step

def prediction_next(const_ar, phi9, phi8, phi7, phi6, phi5, phi4, phi3, phi2, phi1, y_lag1, y_lag2, y_lag3, y_lag4, y_lag5, y_lag6, y_lag7, y_lag8, y_lag9):
    return const_ar + phi9 * y_lag9 + phi8 * y_lag8 + phi7 * y_lag7 + phi6 * y_lag6 + phi5 * y_lag5 + phi4 * y_lag4 + phi3 * y_lag3 + phi2 * y_lag2 + phi1 * y_lag1

def df_forecast_ar9():
    forecast_ar9_step1 = forecasting_ar9(1, df_ar9)
    forecast_ar9_step2 = forecasting_ar9(2, df_ar9)
    forecast_ar9_step3 = forecasting_ar9(3, df_ar9)
    forecast_ar9_step4 = forecasting_ar9(4, df_ar9)

    # Make a dataframe of the forecast of AR 9, name coulms AR(9) step: 1, AR(9) step: 2, AR(9) step: 3, AR(9) step: 4
    df_ar9_results = pd.DataFrame({'AR(9) step: 1': forecast_ar9_step1,
    'AR(9) step: 2': forecast_ar9_step2,
    'AR(9) step: 3': forecast_ar9_step3,
    'AR(9) step: 4': forecast_ar9_step4}, index=df_ar9[start_index:].index)

    # Add real values
    df_ar9_results['True data'] = df_ar9.CPI_growth[start_index:]

    return df_ar9_results
```
